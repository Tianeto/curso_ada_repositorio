```python
titanic = pd.read_csv('titanic.csv')
```

```python
titanic.shape
```

```python
titanic.head()
```

```python
# unique para trazer apenas os valores únicos da coluna e serve para ver as diferente opçoes dentro da coluna

titanic['Sex'].unique()
```

```python
#média do tícket das passagem de homens. Round para arredondar, entre aspas coloca quantas casas, se deixar em branco fica sem nenhuma casa decimal
titanic[titanic['Sex'] == 'male'] ['Fare'].mean().round(2)
```

```python
# passageiro que pagou a maior tarifa
titanic['Fare'].max()
```

```python
# o nome do passageiro que pagou a maior tarifa
titanic[titanic['Fare'] == titanic["Fare"].max()]['Name']
```

```python
# defina a idade média dos sobreviventes
titanic[titanic['Survived'] == 1]['Age'].mean().round()
```

```python
# Determine a classe com mais sobreviventes proporcionalmente
#  1 - quais são as classes?
titanic['Pclass'].unique()
```

```python
total_3_classe = titanic[titanic['Pclass'] == 3].shape[0] # shape vai informar a quantidade de linhas
total_2_classe = titanic[titanic['Pclass'] == 2].shape[0]
total_1_classe = titanic[titanic['Pclass'] == 1].shape[0]
```

```python
#esta na 3 classe e sobreviveu
sobreviveu_3_classe = titanic[(titanic['Pclass'] == 3) & (titanic['Survived'] == 1)].shape[0]

# morreu na 3 classe
morreu_3_classe = titanic[(titanic['Pclass'] == 3) & (titanic['Survived'] == 0)].shape[0]

#proporção dos vivos 3 classe
print('vivos =', sobreviveu_3_classe/total_3_classe)

# proporção mortos 3 classe
print('mortos = ', morreu_3_classe / total_3_classe)
```

```python
#esta na 2 classe e sobrebiveu
sobreviveu_2_classe = titanic[(titanic['Pclass'] == 2) & (titanic['Survived'] == 1)].shape[0]

# morreu na 2 classe
morreu_2_classe = titanic[(titanic['Pclass'] == 2) & (titanic['Survived'] == 0)].shape[0]

#proporção dos vivos 2 classe
print('vivos =', sobreviveu_2_classe / total_2_classe)

# proporção mortos 2 classe
print('mortos = ', morreu_2_classe / total_2_classe)
```

```python
#esta na 1 classe e sobreviveu
sobreviveu_1_classe = titanic[(titanic['Pclass'] == 1) & (titanic['Survived'] == 1)].shape[0]

# morreu na 1 classe
morreu_1_classe = titanic[(titanic['Pclass'] == 1) & (titanic['Survived'] == 0)].shape[0]

#proporção dos vivos 2
# classe
print('vivos =', sobreviveu_1_classe/total_1_classe)

# proporção mortos 2
# classe
print('mortos = ', morreu_1_classe / total_1_classe)
```
